```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
from scipy.stats import boxcox

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python

data = pd.read_csv('/home/kotarum/Ironhack/unit_4/lab-random-variable-distributions/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')  


numerical = data.select_dtypes(include=['float64', 'int64'])
print(numerical.head())
```

```python

for column in numerical.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(numerical[column], kde=True, stat="density", linewidth=0)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python

def plot_transformed_distribution(data, transformation=None):
    for column in data.columns:
        plt.figure(figsize=(10, 6))
        if transformation == 'log':

            transformed_data = np.log(data[column] + 1)
        elif transformation == 'sqrt':

            transformed_data = np.sqrt(data[column])
        elif transformation == 'boxcox':
          
            transformed_data, _ = boxcox(data[column] + 1)
        else:
            transformed_data = data[column]
        
        sns.histplot(transformed_data, kde=True, stat="density", linewidth=0)
        transformation_name = 'Original' if transformation is None else transformation.capitalize()
        plt.title(f'{transformation_name} Distribution of {column}')
        plt.show()


plot_transformed_distribution(numerical)


plot_transformed_distribution(numerical, transformation='log')


plot_transformed_distribution(numerical, transformation='sqrt')

plot_transformed_distribution(numerical, transformation='boxcox')
```
